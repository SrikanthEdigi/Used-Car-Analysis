```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Downloads\car_data.csv")
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(inplace=True,drop=True)
df
```

```python
df.info()
```

```python
df['Test_Drive_Availability'].unique()
```

```python
df['Test_Drive_Availability']=['Yes' if i=='Home Test Drive' else 'No' for i in df['Test_Drive_Availability']]
df
```

```python
df['Test_Drive_Availability'].unique()
```

```python
df.info()
```

```python
df.columns
```

```python
df['Car_Age'] = 2025 - df['Year']
```

```python
df
```

```python
df.info()
```

```python
sns.countplot(df,y="Brand",palette='pastel',hue='Brand')
plt.title('Most Common Car Brands')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.tight_layout()
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Most Common Car Brands.png")
plt.show()
```

```python
fuel_counts=df['Fuel_Type'].value_counts()
plt.title('Fuel Type Preference',loc='right')
plt.pie(fuel_counts,labels=fuel_counts.index,autopct="%1.1f%%",explode=[0.1,0.1,0.4,0.4,0.4,0.4],pctdistance=1.2,labeldistance=1.35)
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Fuel Type Preference.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Car_Age'], bins=15, kde=True,color="Red")
plt.title('Distribution of Car Age')
plt.xlabel('Car Age (Years)')
plt.ylabel('Number of Cars')
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Distribution of Car Age.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Kms_Driven'], bins=30, kde=True)
plt.title('Distribution of Kilometers Driven')
plt.xlabel('KMs Driven')
plt.ylabel('Number of Cars')
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Distribution of Kilometers Driven.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='City', order=df['City'].value_counts().index,palette="muted",hue="City")
plt.xticks(rotation=45,ha="right")
plt.title('Number of Listings by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Number of Listings by City.png")
plt.show()
```

```python
test_drive_counts = df['Test_Drive_Availability'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(test_drive_counts, labels=test_drive_counts.index, autopct='%1.1f%%', startangle=140,shadow=True)
plt.title('Test Drive Availability')
plt.axis('equal')
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Test Drive Availability.png")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, y='Price')
plt.title('Violinplot of Car Prices')
plt.ylabel('Price')
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Violinplot of Car Prices.png")
plt.show()
```

### Bivariate

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Kms_Driven', y='Price',hue='Fuel_Type')
plt.title('KMs Driven vs Car Price')
plt.xlabel('KMs Driven')
plt.ylabel('Price')
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\KMs Driven vs Car Price.png")
plt.show()
```

```python
city_price_avg = df.groupby('City')['Price'].mean().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=city_price_avg.index, y=city_price_avg.values,palette='colorblind',hue=city_price_avg.index)
plt.xticks(rotation=45,ha='right')
plt.title('Average Car Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.tight_layout()
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Average Car Price by Region.png")
plt.show()
```

```python
test_drive_avg_price = df.groupby('Test_Drive_Availability')['Price'].mean()
sns.barplot(x=test_drive_avg_price.index, y=test_drive_avg_price.values,width=0.3,palette='dark',hue=test_drive_avg_price.index)
plt.title('Average Car Price by Test Drive Availability')
plt.xlabel('Test Drive Availability')
plt.ylabel('Average Price')
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Average Car Price by Test Drive Availabilty.png")
plt.show()
```

### MultiVariate 

```python
corr_features = df[['Car_Age', 'Price', 'Kms_Driven']].corr()
sns.heatmap(corr_features, annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Correlation Matrix.png")
plt.show()
```

```python
sns.scatterplot(data=df, x='Car_Age', y='Price', size='Kms_Driven', hue='Fuel_Type', palette='deep')
plt.title('Car Age vs Price (Sized by KMs Driven, Colored by Fuel Type)')
plt.xlabel('Car Age (Years)')
plt.ylabel('Price')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(r"C:\Users\user\Downloads\EDA Plots\Car Age vs Price (Sized by KMs Driven, Colored by Fuel Type).png")
plt.show()
```
