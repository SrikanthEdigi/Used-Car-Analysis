```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.cartrade.com/second-hand/hyderabad/#city=105&sc=-1&so=-1&pn=1'
page=requests.get(url)
```

```python
Soup=BeautifulSoup(page.text)
Soup
```

```python
len(Soup.find_all('div',class_="cr_prc"))
```

```python
Price=[]
for i in Soup.find_all('div',class_="cr_prc"):
    prices=i.text.strip().split()
    price=''.join([i for i in prices[0] if i.isdigit()]) if prices else '0'
    Price.append(int(price))
```

```python
Price
```

```python
Soup.find_all('span',class_="cardTitle font-muli")[0].text.split()[1]
```

```python
data=Soup.find_all('span',class_="cardTitle font-muli")
Brand=[]
for i in data:
    Brand.append(i.text.split()[1])
```

```python
Brand
```

```python
data[0].text.split()[2]
```

```python
Model=[]
for i in data:
    Model.append(i.text.split()[2])
```

```python
Model
```

```python
data[0].text.split()[0]
```

```python
Year=[]
for i in data:
    Year.append(i.text.split()[0])
```

```python
Year
```

```python
data[0].text.split()[1:5]
```

```python
Car_Name=[]
for i in data:
    name_parts=i.text.split()[1:5]
    Car_Name.append(' '.join(name_parts).replace(',',' '))
```

```python
Car_Name
```

```python
Variant=[]
for i in data:
    variant_parts=i.text.split()[3:5]
    Variant.append(' '.join(variant_parts).replace(',',' '))
```

```python
Variant
```

```python
data1=Soup.find_all('span',class_="carInfo")
Kms_Driven=[]
for i in data1:
    kms_parts=i.text.strip().split()
    kms=''.join([i for i in kms_parts[0] if i.isdigit()]) if kms_parts else '0'
    Kms_Driven.append(int(kms))
```

```python
Kms_Driven
```

```python
data2=Soup.find_all('span',class_="carInfo")
Fuel_Type=[]
for i in data2:
    Fuel_Type.append(i.text.split()[3])
```

```python
Fuel_Type
```

```python
data3=Soup.find_all('span',class_="carInfo")
City=[]
for i in data3:
    city_parts=i.text.strip().split()[5:8]
    city_clean=' '.join(city_parts).replace(' ,',',')
    City.append(city_clean)
```

```python
City
```

```python
data4=Soup.find_all('span',class_="tag featured__tag font-muli")
Featured=[]
for i in data4:
    if i in data4:
        Featured.append(i.text.split()[0])
    else:
        Featured.append('Not Featured')
```

```python
Featured
```

```python
data5=Soup.find_all('span',class_="additional-tag-text font-muli")
Test_Drive_Availability=[]
for i in data5:
    test_parts=i.text.split()[0:4]
    Test_Drive_Availability.append(' '.join(test_parts).replace(',',' '))
```

```python
Test_Drive_Availability
```
